```python
from langchain_core.prompts import ChatPromptTemplate, MessagesPlaceholder
```

```python
def create_agent(llm: ChatOllama, tools: list, system_prompt: str):
    # Each worker node will be given a name and some tools.
    prompt = ChatPromptTemplate.from_messages(
        [
            (
                "system",
                system_prompt,
            ),
            MessagesPlaceholder(variable_name="messages"),
            MessagesPlaceholder(variable_name="agent_scratchpad"),
        ]
    )
    agent = create_openai_tools_agent(llm, tools, prompt)
    # print(agent)
    executor = AgentExecutor(agent=agent, tools=tools)
    return executor
```

```python
def agent_node(state, agent, name):
    temp = {'messages':[state['messages'][0]]}
    print('TEMP',temp)
    result = agent.invoke(temp)
    # result = agent.invoke(state)
    return {"messages": [HumanMessage(content=result["output"], name=name)]}
```

```python
class AgentState(TypedDict):
    # The annotation tells the graph that new messages will always
    # be added to the current states
    messages: Annotated[Sequence[BaseMessage], operator.add]
    # The 'next' field indicates where to route to next
    next: str
```

```python
members = ["Recommender", "Summarizer", "Alternatives"]
router_prompt = (
"""
 You are a router for a pharma agent-based system. Your role is to analyze user queries and route them to the most appropriate agent. You must classify each query into one of three categories:

Recommender Agent: For queries where users seek advice or recommendations about medications, treatment options, or specific medical conditions.
Alternatives Agent: For queries where users request alternative medications, treatments, or solutions.
Summarizer Agent: For queries where users ask for a summary or concise explanation of detailed medical information.
Carefully examine the context and intent of the query. Use the examples provided below as guidelines for routing.

Output Format:
Always respond with a JSON object in the following format:
{{'next': '<Agent_Name>'}}
Where <Agent_Name> is one of Recommender, Alternatives, or Summarizer.

For example, if the query should be routed to the Summarizer agent, your response should be:
{{'next': 'Summarizer'}}

Do not deviate from this format.

Examples:
Query: "Can I take Ibuprofen if I have a history of stomach ulcers?"
Routing: {{'next': 'Recommender'}}
Reasoning: This query seeks advice on the suitability of a medication based on a medical condition, which falls under the Recommender agent's domain.

Query: "Are there other medications I can take instead of Ibuprofen for pain relief?"
Routing: {{'next': 'Alternatives'}}
Reasoning: The user is explicitly asking for alternatives to a specific medication, which is handled by the Alternatives agent.

Query: "Can you provide a summary of the side effects of Ibuprofen?"
Routing: {{'next': 'Summarizer'}}
Reasoning: The user is requesting a concise explanation of detailed information, which falls under the Summarizer agent's domain.

Query: "What are the best medications for managing chronic back pain?"
Routing: {{'next': 'Recommender'}}
Reasoning: This is a request for recommendations regarding medications for a specific condition.

Query: "What are the common alternatives to antibiotics for bacterial infections?"
Routing: {{'next': 'Alternatives'}}
Reasoning: The user is seeking alternative treatment options, which is under the Alternatives agent's scope.

Query: "Summarize the differences between generic and brand-name medications."
Routing: {{'next': 'Summarizer'}}
Reasoning: The user is asking for a concise summary of comparative information, appropriate for the Summarizer agent.
"""
)

# Our team supervisor is an LLM node. It just picks the next agent to process
# and decides when the work is completed
# options =  members
```

```python
from langchain_core.output_parsers.openai_functions import JsonOutputFunctionsParser
from langchain_core.utils.function_calling import convert_to_openai_function
from typing import Literal, Union, Sequence, TypedDict, Annotated
from langchain_core.messages.ai import (
    AIMessage,
    AIMessageChunk,
)
import json

options=members


tool_agents=["Search","RAG"]

function_def1 = {
    "name": "route",
    "description": "Select the next role.",
    "parameters": {
        "title": "routeSchema",
        "type": "object",
        "properties": {
            "next": {
                "title": "Next",
                "anyOf": [
                    {"enum": members},
                ],
            }
        },
        "required": ["next"],
    },
}

# function_def2 = {
#     "name": "route",
#     "description": "Select the next role.",
#     "parameters": {
#         "title": "routeSchema",
#         "type": "object",
#         "properties": {
#             "next": {
#                 "title": "Next",
#                 "anyOf": [
#                     {"enum": tool_agents},
#                 ],
#             }
#         },
#         "required": ["next"],
#     },
# }

def route(next: Union[Literal["Recommender"], Literal["Summarizer"], Literal["Alternatives"]]) -> str:
    """
    Select the next role.

    Args:
        next (str): The next role to select. Must be 'Schedule and Venues','Archery', 'Tennis', 'Skateboarding', 'General' or 'Hockey'.

    Returns:
        str: The selected role.

    Raises:
        ValueError: If the provided 'next' value is not 'Schedule and Venues','Archery', 'Tennis', 'Skateboarding', 'General' or 'Hockey'.
    """
    if next not in members:
        raise ValueError(f"Invalid option. Must be one of: {', '.join(members)}")
    return next

# def route2(next: Union[Literal["Search"], Literal["RAG"]]) -> str:
#     """
#     Select the next role.

#     Args:
#         next (str): The next role to select. Must be 'Schedule and Venues','Archery', 'Tennis', 'Skateboarding', 'General' or 'Hockey'.

#     Returns:
#         str: The selected role.

#     Raises:
#         ValueError: If the provided 'next' value is not 'Schedule and Venues','Archery', 'Tennis', 'Skateboarding', 'General' or 'Hockey'.
#     """
#     if next not in tool_agents:
#         raise ValueError(f"Invalid option. Must be one of: {', '.join(tool_agents)}")
#     return next

# Define the supervisor chain
prompt = ChatPromptTemplate.from_messages(
    [
        ("system", router_prompt),
        MessagesPlaceholder(variable_name="messages"),
        (
            "system",
            "Given the conversation above, who should act next?"
            " Or should we FINISH? Select one of: {options}",
        ),
    ]
).partial(options=str(members), members=", ".join(members))

def parse_ai_message(ai_msg: AIMessage) -> dict:
    """
    Parse the content of an AIMessage to extract the 'next' value.

    Args:
        ai_msg (AIMessage): The AI message to parse.

    Returns:
        dict: A dictionary containing the 'next' value.
    """
    try:
        # Try to parse the content as JSON
        # print(ai_msg)
        return json.loads(ai_msg.content)
    except json.JSONDecodeError:
        # If JSON parsing fails, try to evaluate it as a Python literal
        import ast
        try:
            return ast.literal_eval(ai_msg.content)
        except (ValueError, SyntaxError):
            # If both methods fail, return a default value
            # print("Here")
            return {"next": "FINISH"}

router_chain = (
    prompt
    | llm.bind_tools(tools=[function_def1])
)

def router_node(state: AgentState) -> AgentState:
    messages = state["messages"]
    ai_msg = router_chain.invoke({"messages": messages})
    new_messages = list(messages) + [ai_msg]
    # for tool_call in ai_msg.tool_calls:
    #     selected_tool = {"route": function_def}[tool_call["name"].lower()]
    #     #asking LLM to generate function call with arguments
    #     tool_output = selected_tool.invoke(tool_call["args"])
    #     messages.append(ToolMessage(tool_output, tool_call_id=tool_call["id"]))
    ai_msg = router_chain.invoke({"messages": new_messages})

    print(ai_msg)

    # Parse the AI message to extract the 'next' value
    result = parse_ai_message(ai_msg)
    # print(new_messages)

    if "next" in result:
        next_step = result["next"]
        if next_step in members:
            return {"messages": new_messages, "next": next_step}

    # If parsing fails or 'next' is not a valid option, default to "supervisor"
    return {"messages": new_messages, "next": "Summarizer"}

workflow = StateGraph(AgentState)

# Add nodes
workflow.add_node("Recommender", recommendor_node)
workflow.add_node("Summarizer",summarizer_agent)
workflow.add_node("Alternatives",alternative_agent)
# workflow.add_node("Search", other_node)
# workflow.add_node("RAG", other_node)
# workflow.add_node("Hockey", hockey_node)



workflow.add_node("Router", router_node)

# Set entry point
workflow.set_entry_point("Router")

# for i in members:
#   workflow.add_edge(i,"Router")
#   workflow.add_edge("Router",i)


# main_agents=[]

# Add conditional edges
conditional_map = {k: k for k in options}
# conditional_map["FINISH"] = END

workflow.add_conditional_edges(
    "Router",
    lambda x: x["next"],
    conditional_map
)

# for i in members:
#   workflow.add_edge(i,"Router")

# for member in members:
#     # We want our workers to ALWAYS "report back" to the supervisor when done
#     workflow.add_edge(member, "supervisor")
# Add edges from other nodes back to supervisor
# workflow.add_edge("Schedule", "supervisor")

# Compile the graph
graph = workflow.compile()
```
